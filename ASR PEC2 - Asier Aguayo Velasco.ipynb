```python
columns = ['userID', 'friendID']
dfEdges = pd.DataFrame()
dfFriends = df.drop_duplicates('userID')

rows = []

for i in range(0,len(dfFriends)):
    p = dfFriends.iloc[i]
    next_cursor = -1
    
    while next_cursor != 0:
        resultFriends = twitter.get_friends_ids(user_id = p['userID'],
                                               count = p['friends'],
                                               cursor = next_cursor)
    
        next_cursor = resultFriends['next_cursor']
        for friend in resultFriends['ids']:
            friendData = {
                'userID' : p['userID'],
                'friendID' : friend,
            }
            rows.append(friendData)
        
    i += 1

dfEdges = pd.DataFrame(rows, columns=columns)
dfEdges
```
